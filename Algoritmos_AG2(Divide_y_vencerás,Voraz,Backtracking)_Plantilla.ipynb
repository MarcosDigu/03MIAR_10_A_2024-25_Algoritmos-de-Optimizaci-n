## Práctica Individual: Encontrar los dos puntos más cercanos

### Problema
El objetivo es encontrar los dos puntos más cercanos en diferentes dimensiones.

- Dividir el conjunto de puntos en subproblemas más pequeños, resolver cada subproblema y combinar los resultados.
- **Calcular complejidad:** Determinar si es más eficiente que el enfoque de fuerza bruta.

```python
# 1.1. Fuerza bruta

import random
import numpy as np
import time

lista_1d = [round(random.uniform(0, 100), 3) for _ in range(10**4)]
print(f'Lista 1D: {lista_1d}')
print(' ')

def fb_1d(lista_1d):

    distancia_minima = np.inf
    elem1, elem2 = 0, 0

    for num1 in lista_1d:
        
        for num2 in [num for num in lista_1d if num != num1]:

            distancia = abs(num1 - num2)

            if distancia < distancia_minima:

                distancia_minima, elem1, elem2 = round(distancia, 3), num1, num2
            
    return distancia_minima, elem1, elem2

start_time = time.time()
distancia_minima, elem1, elem2 = fb_1d(lista_1d=lista_1d)
end_time = time.time()

print('tiempo de ejecución', round(end_time - start_time, 6))
print('distancia_minima', distancia_minima)
print('elem1', elem1)
print('elem2', elem2)
```

## Análisis de la complejidad computacional del algoritmo de Fuerza Bruta

### Descripción del algoritmo
El algoritmo busca los dos puntos más cercanos en una lista unidimensional `lista_1d` utilizando el enfoque de fuerza bruta. Esto implica calcular la distancia entre **todas las posibles parejas de puntos** y mantener actualizada la distancia mínima encontrada hasta el momento.

### Pasos principales del algoritmo
1. **Recorrer todos los elementos de la lista (`num1`)**:  
   Esto genera el primer bucle que tiene una complejidad de \(O(n)\), donde \(n\) es el tamaño de la lista.
   
2. **Para cada elemento, comparar con los demás (`num2`)**:  
   El segundo bucle también recorre \(n-1\) elementos en el peor caso (todas las parejas excepto el elemento actual). Este bucle anidado tiene una complejidad de \(O(n)\).

3. **Calcular la distancia y actualizar la mínima si es necesario**:  
   Estas son operaciones constantes, con una complejidad de \(O(1)\) por cada comparación.

### Complejidad total
El algoritmo realiza $(n \cdot (n-1))$ iteraciones, una para cada pareja de puntos. Esto da lugar a una complejidad cuadrática: $O(n^2)$.

Este comportamiento se debe a los dos bucles anidados que comparan cada punto con los demás.

### Escalabilidad del algoritmo
- Para listas pequeñas, este enfoque es aceptable, ya que $O(n^2)$ es manejable.
- Sin embargo, para listas más grandes $(n > 10^4)$, el tiempo de ejecución crece rápidamente, lo que lo hace ineficiente en comparación con métodos más avanzados como **divide y vencerás**.


```python
# 1.2. Divide y vencerás

import random
import numpy as np
import time

lista_1d = [round(random.uniform(0, 100), 3) for _ in range(5)]
print(f'Lista 1D: {lista_1d}')
print(' ')

def dv_1d(points):
    """
    Encuentra los dos puntos más cercanos en una lista de puntos 1D usando divide y vencerás.
    """
    # Paso 1: Caso base
    if len(points) <= 3:
        # Si hay 3 o menos puntos, resolver con fuerza bruta
        min_distance = float('inf')
        p1, p2 = None, None
        for i in range(len(points)):
            for j in range(i + 1, len(points)):
                distance = abs(points[i] - points[j])
                if distance < min_distance:
                    min_distance = distance
                    p1, p2 = points[i], points[j]
        return min_distance, p1, p2

    # Paso 2: Dividir
    mid = len(points) // 2
    left_half = points[:mid]
    right_half = points[mid:]

    # Resolver recursivamente en cada mitad
    min_left, p1_left, p2_left = dv_1d(left_half)
    min_right, p1_right, p2_right = dv_1d(right_half)

    print('left_half', left_half)
    print('min_left', min_left)
    print('p1_left', p1_left)
    print('p2_left', p2_left)
    print('--------')    
    print('right_half', right_half)
    print('min_right', min_right)
    print('p1_right', p1_right)
    print('p2_right', p2_right)
    print(' ')

    # Paso 3: Combinar
    # Encuentra el mínimo entre las dos mitades
    delta = min(min_left, min_right)
    closest_pair = (p1_left, p2_left) if min_left < min_right else (p1_right, p2_right)

    print('delta', delta)
    print('closest_pair', closest_pair)

    # Considerar la franja (puntos en la frontera)
    mid_point = points[mid]
    candidates = [point for point in points if abs(point - mid_point) < delta]

    print('mid_point', mid_point)
    print('candidates', candidates)

    # Comparar los puntos en la franja
    candidates.sort()  # Asegurarse de que estén ordenados para comparar consecutivos
    for i in range(len(candidates)):
        for j in range(i + 1, len(candidates)):
            if abs(candidates[j] - candidates[i]) >= delta:
                break
            distance = abs(candidates[j] - candidates[i])
            if distance < delta:
                delta = distance
                closest_pair = (candidates[i], candidates[j])

    return delta, closest_pair[0], closest_pair[1]

# Ejecutar algoritmo
t0 = time.time()
min_distance, point1, point2 = dv_1d(lista_1d)
t1 = time.time()

print('tiempo de ejecución', round(t1 - t0, 6))
print(f"Distancia mínima: {min_distance}")
print(f"Puntos más cercanos: {point1}, {point2}")
```

## Análisis de la complejidad computacional del algoritmo de Divide y Vencerás

### Descripción del algoritmo
El algoritmo encuentra los dos puntos más cercanos en una lista de puntos 1D usando el enfoque de **divide y vencerás**. Divide la lista en dos mitades recursivamente, encuentra los pares más cercanos en cada mitad y combina las soluciones, evaluando posibles pares que cruzan la división.

---

### Complejidad por pasos

#### 1. **Caso base**:
- Si el número de puntos es $( \leq 3 )$, el algoritmo usa fuerza bruta para calcular todas las distancias posibles.
- **Complejidad**: $( O(1) )$ en promedio porque no afecta significativamente el tamaño global de la lista.

#### 2. **División**:
- El algoritmo divide la lista en dos mitades:
  - Izquierda: `points[:mid]`
  - Derecha: `points[mid:]`
- **Complejidad**: $( O(1) )$ porque las operaciones de división de listas son rápidas y no dependen del tamaño del problema.

#### 3. **Resolución recursiva**:
- Se resuelve el problema recursivamente en ambas mitades.
- Cada vez que dividimos, reducimos el tamaño de la lista a la mitad: $T(n) = 2 \cdot T(n/2) + O(n)$. El término $( O(n) )$ proviene del paso de combinación.

#### 4. **Combinación (la franja)**:
- Una vez resuelto el problema en las mitades, el algoritmo combina los resultados:
  - Determina la distancia mínima $\delta$ entre las mitades.
  - Evalúa los puntos en una franja de ancho $2 \cdot \delta$ alrededor del punto medio.
  - La franja se recorre de forma eficiente porque compara solo los puntos cercanos.

- **Complejidad del paso de combinación**: $ O(n) $, ya que cada punto se compara con un número limitado de puntos cercanos en la franja.

---

### Complejidad total
### Resolución de la recurrencia del tiempo

La recurrencia del tiempo es:  
$T(n) = 2 \cdot T(n/2) + O(n)$  
Esto corresponde al costo de resolver dos subproblemas de tamaño $n/2$ y combinar los resultados en $O(n)$.

Usamos el **Teorema Maestro** para resolver la recurrencia:

1. **Forma estándar del Teorema Maestro**:  
   $T(n) = a \cdot T(n/b) + O(n^d)$  
   Donde:
   - $a = 2$: número de subproblemas.
   - $b = 2$: factor de reducción del tamaño del problema.
   - $f(n) = O(n)$: costo de la combinación.

2. **Cálculo de $p = \log_b(a)$**:  
   $p = \log_2(2) = 1$

3. **Comparación entre $n^p$ y $f(n)$**:  
   - $n^p = n^1 = O(n)$  
   - $f(n) = O(n)$  

   Como $n^p$ y $f(n)$ son iguales ($p = d$), estamos en el caso **2 del Teorema Maestro**.

4. **Resultado del caso 2**:  
   Cuando $p = d$, la solución es:  
   $T(n) = O(n \log n)$

Por lo tanto, la complejidad total del algoritmo es:  
$T(n) = O(n \log n)$

---

### Comparación con fuerza bruta
- **Fuerza bruta**: $ O(n^2) $ porque compara todas las parejas posibles.
- **Divide y vencerás**: $ O(n \log n) $, lo que lo hace mucho más eficiente para listas grandes.

---

### Conclusión
El enfoque de **divide y vencerás** reduce significativamente la complejidad computacional del problema de encontrar los puntos más cercanos. Aunque el paso de combinación puede parecer costoso, el uso de la franja limita las comparaciones, manteniendo la eficiencia en $ O(n \log n) $.

---

### ¿Se puede mejorar el algoritmo actual de $O(n \log n)$?

Sinceramente, no creo que se pueda mejorar más allá de $O(n \log n)$. La razón es que este tipo de problema ya implica operaciones de ordenamiento o comparaciones que, en el peor caso, requieren al menos $O(n \log n)$. Básicamente, estamos atrapados en este límite porque el proceso de dividir y combinar depende de recorrer los elementos y comparar pares.

Eso sí, aunque no se pueda mejorar la complejidad asintótica, siempre se pueden optimizar los detalles. Por ejemplo, reducir la constante oculta o simplificar algunos cálculos puede hacer que funcione más rápido en la práctica, pero en términos generales, ya estamos en el mejor nivel que podemos alcanzar para este problema.


## Extender el algoritmo a 2D

```python
import random
import time

# Generar una lista de puntos aleatorios en 2D
lista_2d = [(round(random.uniform(0, 100), 3), round(random.uniform(0, 100), 3)) for _ in range(10)]
print(f'Lista 2D: {lista_2d}\n')

def dv_2d(points):
    """
    Encuentra los dos puntos más cercanos en una lista de puntos 2D usando divide y vencerás.
    """
    # Paso 1: Caso base
    if len(points) <= 3:
        # Si hay 3 o menos puntos, resolver con fuerza bruta
        min_distance = float('inf')
        p1, p2 = None, None
        for i in range(len(points)):
            for j in range(i + 1, len(points)):
                distance = ((points[i][0] - points[j][0]) ** 2 + (points[i][1] - points[j][1]) ** 2) ** 0.5
                if distance < min_distance:
                    min_distance = distance
                    p1, p2 = points[i], points[j]
        return min_distance, p1, p2

    # Paso 2: Dividir
    points.sort(key=lambda x: x[0])  # Ordenar por la coordenada x
    mid = len(points) // 2
    mid_x = points[mid][0]
    left_half = points[:mid]
    right_half = points[mid:]

    # Resolver recursivamente en cada mitad
    min_left, p1_left, p2_left = dv_2d(left_half)
    min_right, p1_right, p2_right = dv_2d(right_half)

    # Encuentra el mínimo entre las dos mitades
    delta = min(min_left, min_right)
    closest_pair = (p1_left, p2_left) if min_left < min_right else (p1_right, p2_right)

    # Paso 3: Considerar la franja (puntos en la frontera)
    strip = [point for point in points if abs(point[0] - mid_x) < delta]

    # Ordenar los puntos de la franja por la coordenada y
    strip.sort(key=lambda x: x[1])

    # Comparar los puntos en la franja
    for i in range(len(strip)):
        for j in range(i + 1, len(strip)):
            if strip[j][1] - strip[i][1] >= delta:
                break
            distance = ((strip[i][0] - strip[j][0]) ** 2 + (strip[i][1] - strip[j][1]) ** 2) ** 0.5
            if distance < delta:
                delta = distance
                closest_pair = (strip[i], strip[j])

    return delta, closest_pair[0], closest_pair[1]

# Ejecutar algoritmo
t0 = time.time()
min_distance, point1, point2 = dv_2d(lista_2d)
t1 = time.time()

print('tiempo de ejecución:', round(t1 - t0, 6))
print(f"Distancia mínima: {min_distance}")
print(f"Puntos más cercanos: {point1}, {point2}")
```

```python
# Generar una lista de puntos aleatorios en 3D
lista_3d = [(round(random.uniform(0, 100), 3), round(random.uniform(0, 100), 3), round(random.uniform(0, 100), 3)) for _ in range(10)]
print(f'Lista 3D: {lista_3d}\n')

def dv_3d(points):
    """
    Encuentra los dos puntos más cercanos en una lista de puntos 3D usando divide y vencerás.
    """
    # Paso 1: Caso base
    if len(points) <= 3:
        # Si hay 3 o menos puntos, resolver con fuerza bruta
        min_distance = float('inf')
        p1, p2 = None, None
        for i in range(len(points)):
            for j in range(i + 1, len(points)):
                distance = ((points[i][0] - points[j][0]) ** 2 +
                            (points[i][1] - points[j][1]) ** 2 +
                            (points[i][2] - points[j][2]) ** 2) ** 0.5
                if distance < min_distance:
                    min_distance = distance
                    p1, p2 = points[i], points[j]
        return min_distance, p1, p2

    # Paso 2: Dividir
    points.sort(key=lambda x: x[0])  # Ordenar por la coordenada x
    mid = len(points) // 2
    mid_x = points[mid][0]
    left_half = points[:mid]
    right_half = points[mid:]

    # Resolver recursivamente en cada mitad
    min_left, p1_left, p2_left = dv_3d(left_half)
    min_right, p1_right, p2_right = dv_3d(right_half)

    # Encuentra el mínimo entre las dos mitades
    delta = min(min_left, min_right)
    closest_pair = (p1_left, p2_left) if min_left < min_right else (p1_right, p2_right)

    # Paso 3: Considerar la franja (puntos en la frontera)
    strip = [point for point in points if abs(point[0] - mid_x) < delta]

    # Ordenar los puntos de la franja por la coordenada y
    strip.sort(key=lambda x: x[1])

    # Comparar los puntos en la franja
    for i in range(len(strip)):
        for j in range(i + 1, len(strip)):
            if strip[j][1] - strip[i][1] >= delta:
                break
            distance = ((strip[i][0] - strip[j][0]) ** 2 +
                        (strip[i][1] - strip[j][1]) ** 2 +
                        (strip[i][2] - strip[j][2]) ** 2) ** 0.5
            if distance < delta:
                delta = distance
                closest_pair = (strip[i], strip[j])

    return delta, closest_pair[0], closest_pair[1]

# Ejecutar algoritmo
t0 = time.time()
min_distance, point1, point2 = dv_3d(lista_3d)
t1 = time.time()

print('tiempo de ejecución:', round(t1 - t0, 6))
print(f"Distancia mínima: {min_distance}")
print(f"Puntos más cercanos: {point1}, {point2}")
```

### Ajustes necesarios en el algoritmo de Divide y Vencerás al aumentar la dimensión

Al extender el algoritmo de **divide y vencerás** de 1D a 2D, 3D o dimensiones superiores, los siguientes ajustes son necesarios:

---

#### **1. Cálculo de la distancia**
- **1D**: La distancia es la diferencia absoluta entre dos números:  

  $|x_1 - x_2|$

- **2D**: Se usa la fórmula de distancia euclidiana en el plano:  

  $\sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2}$

- **3D**: Se extiende a tres coordenadas:  

  $\sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2 + (z_1 - z_2)^2}$

- En dimensiones superiores (\(n\)-dimensional):  

  $\sqrt{\sum_{i=1}^n (x_i - x'_i)^2}$

---

#### **2. División de los puntos**
- En todas las dimensiones, los puntos se dividen en dos mitades en función de una sola coordenada.  
  - **1D**: Los puntos se dividen por su valor en el eje **x**.
  - **2D y 3D**: Los puntos se dividen por la coordenada **x**. El resto de las coordenadas (**y**, **z**, etc.) no afectan la división.

---

#### **3. Franja alrededor de la línea divisoria**
- La franja se define por la distancia mínima $\delta$ encontrada en las mitades:
  - **1D**: La franja contiene puntos cuya diferencia en **x** con respecto al punto medio es menor que $\delta$.
  - **2D**: Los puntos en la franja son aquellos cuya diferencia en **x** es menor que $\delta$. Estos puntos se ordenan por la coordenada **y** para facilitar comparaciones eficientes.
  - **3D**: Similar a 2D, pero los puntos se ordenan por una coordenada adicional (**y**) y se comparan basándose en las coordenadas restantes (**y** y **z**).

---

#### **4. Comparaciones dentro de la franja**
- **1D**: Comparar cada punto de la franja con el siguiente.
- **2D**: Comparar cada punto en la franja con los puntos cercanos en la coordenada **y** (máximo de 6-7 puntos gracias a propiedades geométricas).
- **3D**: Comparar cada punto en la franja con los puntos cercanos en **y** y **z**, manteniendo la eficiencia.

---

### **Resumen de ajustes**
1. Se adapta la fórmula de distancia al número de dimensiones.
2. La división sigue basándose en la coordenada **x**, sin importar la dimensión.
3. La franja y las comparaciones internas se manejan considerando las coordenadas adicionales, ordenando y limitando el número de comparaciones gracias a la geometría del espacio.
